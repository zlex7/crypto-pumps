```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import os
```

```python
# urls = {'Towards Data Science': 'https://towardsdatascience.com/archive/{0}/{1:02d}/{2:02d}'}
urls = {'The Coinbase Blog': 'https://blog.coinbase.com/archive/{0}/{1:02d}/{2:02d}'}
```

```python
def is_leap(year):
    if year % 4 != 0:
        return False
    elif year % 100 != 0:
        return True
    elif year % 400 != 0:
        return False
    else:
        return True
    
def convert_day(day, year):
    month_days = [31, 29 if is_leap(year) else 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    m = 0
    d = 0
    while day > 0:
        d = day
        day -= month_days[m]
        m += 1
    return (m, d)
```

```python
import time
import numpy as np
data = []
article_id = 0
i = 0
# n = len(selected_days)
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}

import requests
from lxml.html import fromstring
def get_proxies():
    url = 'https://free-proxy-list.net/'
    response = requests.get(url)
    parser = fromstring(response.text)
    proxies = set()
    for i in parser.xpath('//tbody/tr')[:10]:
        if i.xpath('.//td[7][contains(text(),"yes")]'):
            #Grabbing IP and corresponding PORT
            proxy = ":".join([i.xpath('.//td[1]/text()')[0], i.xpath('.//td[2]/text()')[0]])
            proxies.add(proxy)
    return list(proxies)

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.chrome.options import Options

options = Options()
# options.headless = True
driver = webdriver.Chrome(executable_path='./chromedriver',options=options)
# driver = webdriver.Chrome()

proxies = get_proxies()
print(proxies)
# print(len(selected_days))
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2019, 1, 1)
end_date = date(2021, 3, 4)
for dateobj in daterange(start_date, end_date):
    print(dateobj)
    ind = np.random.randint(0,len(proxies))
    proxy = proxies[ind]
#     time.sleep(np.random.randint(1, 15))
    i += 1
#     month, day = convert_day(d, year)
    year = dateobj.year
    month = dateobj.month
    day = dateobj.day
    date = '{0}-{1:02d}-{2:02d}'.format(year, month, day)
    #print(f'{i} / {n} ; {date}')
    for publication, url in urls.items():
        print(url.format(year,month,day))
        driver.get(url.format(year,month,day))
#         response = driver.page_source
#         response = requests.get(url.format(year, month, day), allow_redirects=True,headers=headers,proxies={"http": 'http://%s' % proxy, "https": 'https://%s' %proxy})
        #print 
        if not driver.current_url.startswith(url.format(year,month,day)):
#             print('continuing whoops')
            continue
        page = driver.page_source
        print(page)
#         page = response.content
#         print(response.status_code)
#         print(page)
        soup = BeautifulSoup(page, 'html.parser')

        articles = soup.find_all("div", class_="postArticle postArticle--short js-postArticle js-trackPostPresentation js-trackPostScrolls")
#         print(articles)
        for article in articles:
            title = article.find("h3", class_="graf--title")
            print(title)
            if title is None:
                continue
            title = title.contents[0]
            article_id += 1
            subtitle = article.find("h4", class_="graf--subtitle")
            subtitle = subtitle.contents[0] if subtitle is not None else ''
            article_url = article.find_all("a")[3]['href'].split('?')[0]

            data.append([article_id, article_url, title, subtitle, publication, date])
```

```python
medium_df = pd.DataFrame(data, columns=['id', 'url', 'title', 'subtitle', 'publication', 'date'])
```
